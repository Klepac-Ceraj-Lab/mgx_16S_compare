```python
with open("fastq_filenames", "r") as f:
    fastq=[i for line in f for i in line.split(',')]
```

```python
# remove green lake samples
fastq = [sample for sample in fastq if sample.startswith("C")]
fastq.sort()
```

```python
fastq = [s.replace('\n', '') for s in fastq]
```

```python
len(fastq)
```

```python
keep = []
subjects = []
reads = []
```

```python
for sample in fastq:
    subject = sample.split("-")[0]
    read = sample.split("_")[0]
    if read in reads:
        keep.append(sample)
    elif subject not in subjects:
        reads.append(read)
        keep.append(sample)
        subjects.append(subject)
```

```python
with open('cross-sec-fastq.txt', 'w') as f:
    for item in keep:
        f.write("%s\n" % item)
```

```python
with open('cross-sec-samples.txt', 'w') as f:
    for item in reads:
        f.write("%s\n" % item)
```
